## Data Cleansing

#### We are removing columns that contain more than 50 percent of null values because filling these values make dataset bias and give wrong results

```python
null_50 = pd.DataFrame(null_50)
list = null_50.index
list
```

```python
df = df.drop(list, axis = 1)
```

```python
df.head()
```

### Checking dataset unique value

```python
for i in df.columns:
    print(i, df[i].unique().shape)  
```

#### From above observation we find that ['HCP_Agreement','Responder '] these dataset didn't provide any information throughout the dataset as these dataset columns having only one unique value hence we are removing it.

```python
df = df.drop(['HCP_Agreement','Responder'], axis = 1)
```

```python
df.head()
```

```python
j = []  # unique columns which having less than 10 unique values
for i in df.columns:
    if (df[i].nunique()<=10):
        j.append(i)
        
for i in j:
    print("--------------"+i+"---------------------",'\n' ,df[i].value_counts(), "\n")  # checking unique counts
```

### Filling missing values

```python
print("No. of columns containing null values")
null_col = df[df.columns[df.isna().any()]].columns
null_col
```

```python
for i in null_col:
    df[i] = df[i].fillna(df[i].mode()[0]) 
```

```python
print("No. of columns containing null values")
null_col = df[df.columns[df.isna().any()]].columns
null_col
```
