```python
# importing important libraris
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
import warnings
import time
warnings.filterwarnings("ignore")
import numpy as np
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
from selenium.webdriver.chrome.options import Options
options = webdriver.ChromeOptions()
options. add_experimental_option("detach", True)
#timestamp
import datetime;
```

```python
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(2)
driver.maximize_window()
driver.get("https://www.google.com/")
time.sleep(2)
delay = 10
```

```python
search = driver.find_element_by_xpath('//input[@class = "gLFyf"]')
```

```python
search.send_keys("Samsung mobiles")
search.send_keys(Keys.ENTER)
```

```python
try:
    search1 = driver.find_elements_by_xpath("//div[@class = 'hdtb-mitem']")
    for i in search1:
        if i.text == "Shopping":
            i.click()
    
except StaleElementReferenceException:
     pass
```

```python
items = []
item = driver.find_elements_by_xpath("//div[@class='KZmu8e']")
for i in item:
    items.append(i.text)
```

```python
cell = []
for i in items:
    cell.append(i.split("\n")[0])

prices = []
for i in items:
    prices.append(i.split("\n")[1])

stores = []
for i in items:
    stores.append(i.split("\n")[2])
    
delivery = []
for i in items:
    delivery.append(i.split("\n")[3])

```

```python
url = driver.find_elements_by_xpath("//a[@class = 'shntl sh-np__click-target']")
product_url = []
for i in url:
    product_url.append(i.get_attribute("href"))
```

```python
image_urls = driver.find_elements_by_xpath("//div[@class='SirUVb sh-img__image']/img")

img_url = []
for i in image_urls:
    img_url.append(i.get_attribute("src"))
```

```python
len(cell), len(prices), len(delivery), len(product_url), len(img_url)
```

```python
ct = datetime.datetime.now()
```

```python
df = pd.DataFrame({"Mobile": cell,"Image_Url":img_url, "Prices": prices,"Refurbished": "--", "Stores": stores, "Delivery":delivery, "Timestamp": ct, "Product_Url":product_url})
```

```python
for i in df[df.Prices.str.contains(pat = str('refurbished'), case = False) & df.Prices.str.contains(pat = str('₹'), case = False)].index:
    df['Refurbished'][i] = df['Refurbished'][i].replace("--",'Refurbished_mobile')
    
```

```python
prices = []
for i in items:
    prices.append(i.split("\n")[1].replace("refurbished",""))
df['Prices'] = prices
```
